```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('Salary.csv')
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=df)
plt.title('Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
X = df['YearsExperience'].values.reshape(-1, 1)
y = df['Salary']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Salary vs. Years of Experience with Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared (R2): {r2:.2f}')
```

## Summary:

### Data Analysis Key Findings

*   A scatter plot of 'YearsExperience' vs. 'Salary' shows a strong positive linear relationship.
*   The Mean Squared Error (MSE) of the linear regression model is 35006768.19.
*   The R-squared (R2) score of the model is 0.97, indicating that the model explains 97% of the variance in the salary data.
